```python
with open('moex_data/tickers.txt') as f:
    tickers = f.readlines()
tickers = [t.strip() for t in tickers]
tickers
```

```python
for ticker in tickers:
    print(ticker)
    df_deals = pd.read_feather(f"moex_data/{ticker}_deal.feather")
    df_ticks = pd.read_feather(f"moex_data/{ticker}_tick.feather")
    
    df_deals_prep = prep.preprocess_deals_data(df_deals)
    df_ticks_prep = prep.preprocess_lob_data(df_ticks)
    
    prep.add_lob_prices(df_deals_prep, df_ticks_prep).to_feather(f'data/{ticker}.feather')
```
